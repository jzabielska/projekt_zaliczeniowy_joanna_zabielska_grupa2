1. Zamie≈Ñ skr√≥ty nazw dni tygodnia na polskie odpowiedniki (pe≈Çne nazwy dni tygodnia) - kolumna created_at

```python
def dni_tyg(text):
  dni_tyg = text.replace("Mon", "Poniedzia≈Çek").replace("Tue", "Wtorek").replace("Wed", "≈öroda").replace("Thr", "Czwartek").replace("Fri", "PiƒÖtek").replace("Sat", "Sobota").replace("Sun", "Niedziela")
  return dni_tyg
```

7. Podaj, w kt√≥ry dzie≈Ñ tygodnia najczƒô≈õciej by≈Çy publikowane tweety z datasetu

```python
dni = ['Poniedzia≈Çek', 'Wtorek', '≈öroda', 'Czwartek', 'PiƒÖtek', 'Sobota', 'Niedziela']

counts = {word: 0 for word in dni}
for index, row in df.iterrows():
    for word in dni:
        if word in row["created_at"]:
            counts[word] += 1

print("Liczba wystƒÖpie≈Ñ s≈Ç√≥w:")
for word, count in counts.items():
    print("- {}: {}".format(word, count))

most_common_word = max(counts, key=counts.get)

print("Dzie≈Ñ tygodnia, w kt√≥ry najczƒô≈õciej by≈Çy publikowane tweety: ", most_common_word)
```

**Czƒô≈õƒá 3 - przetwarzanie jƒôzyka naturalnego**

1. Rozpoznaj osoby (persName) i przeka≈º je do kolumny o nazwie persons

```python
def extract_persons(text):
    doc = nlp(text)
    persons = [ent.text for ent in doc.ents if ent.label_ == "persName"]
    return persons

df['persons'] = df['text'].apply(extract_persons)
```

```python
df.persons
```

2. Rozpoznaj miejsca (placeName) i przeka≈º je do kolumny o nazwie places

```python
def extract_places(text):
    doc = nlp(text)
    places = [ent.text for ent in doc.ents if ent.label_ == "placeName"]
    return places

df['places'] = df['text'].apply(extract_places)
```

```python
df.places
```

3. Rozpoznaj organizacje (orgName) i przeka≈º je do kolumny o nazwie organizations

```python
def extract_organizations(text):
    doc = nlp(text)
    organizations = [ent.text for ent in doc.ents if ent.label_ == "orgName"]
    return organizations

df['organizations'] = df['text'].apply(extract_organizations)
```

**Czƒô≈õƒá 4 - rozwiƒÖzywanie problem√≥w/umiejƒôtno≈õƒá interpretowania dokumentacji**

WykorzystujƒÖc pakiet matplotlib wykonaj wykres ilustrujƒÖcy liczbƒô tweet√≥w per dzie≈Ñ tygodnia

Wersja 1 (b≈ÇƒÖd w postaci pustego wykresu)

```python
import matplotlib.pyplot as plt

dni = ['Poniedzia≈Çek', 'Wtorek', '≈öroda', 'Czwartek', 'PiƒÖtek', 'Sobota', 'Niedziela']

licz_dni = {}
for dni in dni:
    licz_dni[dni] = len(df[df['created_at'] == dni])

dni_list = list(licz_dni.keys())
tweety_list = list(licz_dni.values())

plt.bar(dni_list, tweety_list)
plt.xlabel("Dzie≈Ñ tygodnia")
plt.ylabel("Liczba tweet√≥w")
plt.title("Liczba tweet√≥w per dzie≈Ñ tygodnia")
plt.xticks(rotation=45)
plt.show()
```

Wersja 2 (b≈Çƒôdny wykres)

```python
dni = ['Poniedzia≈Çek', 'Wtorek', '≈öroda', 'Czwartek', 'PiƒÖtek', 'Sobota', 'Niedziela']

df["count"] = df["created_at"].apply(lambda x: sum([1 for w in dni if w in x.lower()]))
```

```python
grouped = df.groupby("created_at").sum()
```

```python
grouped.plot(kind='bar', y='count')
plt.title('Liczba tweet√≥w wzglƒôdem dni tygodnia')
plt.xlabel('Dzie≈Ñ tygodnia')
plt.ylabel('Liczba tweet√≥w')

plt.show()
```

Wersja 3 (z≈Çy format daty w kolumnie)

```python
import matplotlib.pyplot as plt

df['created_at'] = pd.to_datetime(df['created_at'])
df['day_of_week'] = df['created_at'].dt.day_name()

day_counts = df.groupby('day_of_week')['user_name'].count()

day_counts.plot(kind='bar')
plt.xlabel('Dzie≈Ñ tygodnia')
plt.ylabel('Liczba tweet√≥w')
plt.title('Liczba tweet√≥w wzglƒôdem dni tygodnia')
plt.show()
```
